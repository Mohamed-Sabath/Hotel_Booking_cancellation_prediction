## It is an imbalanced Dataset

```python
from sklearn.utils import resample

# Undersample the majority class
not_canceled = Data[Data['booking status'] == 'Not_Canceled']
canceled = Data[Data['booking status'] == 'Canceled']

not_canceled_sampled = resample(not_canceled, replace=False, n_samples=len(canceled), random_state=42)

# Combine the sampled majority class with the minority class
balanced_data = pd.concat([not_canceled_sampled, canceled])
```

```python
# One hot encoding

import pandas as pd

df = balanced_data.copy()

# Select the columns you want to one-hot encode
columns_to_encode = ['market segment type']

# Use pd.get_dummies to perform one-hot encoding
df = pd.get_dummies(balanced_data, columns=columns_to_encode)

# Now, 'df' contains the original columns along with the one-hot encoded columns
```

```python
df.head()
```

##  Splitting the Data

```python
x = df.drop('booking status', axis=1)
y = df['booking status']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.75,random_state=0)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#Scaling the values
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()

x_train = mm.fit_transform(x_train)
x_test = mm.transform(x_test)
```

```python
x_train
```

## Training the Models

```python
from sklearn.linear_model import LogisticRegression
# creating a model
model = LogisticRegression()

# feeding the training data to the model
model.fit(x_train, y_train)
```

```python
#Checking the accuracy
print(round(model.score(x_test,y_test)*100,2),'%')
```

```python
from sklearn.tree import DecisionTreeClassifier
modelDT = DecisionTreeClassifier()

# feeding the training data to the model
modelDT.fit(x_train, y_train)

#Checking the accuracy
print(round(modelDT.score(x_test,y_test)*100,2),'%')
```
